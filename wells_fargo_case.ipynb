### Import Data and Storage in Hash Table

```python
import numpy as np
import pandas as pd
import os
import re
```

```python
path = os.getcwd()
folder = "output_wells_fargo_documentai"
folder_file = "file_0"
files = os.listdir(os.path.join(folder, folder_file))
```

```python
dic_data = {}
for file in files:
    data = pd.read_csv(os.path.join(path, folder, folder_file, file))
    p = "(pg\d\w+)"
    regex_p= re.compile(p)
    dic_key = regex_p.search(file).group()
    dic_data[dic_key] = data
```

```python
for key in dic_data.keys():
    print(key)
```

### Separate used table and Storage in Another Dic

```python
tables = {}
for key in dic_data.keys():
    if (len(dic_data[key].columns) > 3) or (len(dic_data[key].columns) == 3 and "Ending daily balance" in dic_data[key].columns):
        tables[key] = dic_data[key] 
    # elif len(dic_data[key].columns) == 3 and "Amount" in dic_data[key].columns:
    #     tables[key] = dic_data[key]
```

### Preprocessing tables 

```python
tables_2 = {}
for key in tables.keys():
    if len(tables[key].columns) == 4:
        
        table = tables[key]
        
        patron = "(\d*|\d{1,3}(,\d{3})*)(\.\d+)"
        regex_number_extractor = re.compile(patron)
        table["Additions"] = table["Deposits/ Additions Description"].apply(lambda x: regex_number_extractor.search(x))
        table["Additions"] = table["Additions"].apply(lambda x: x if x is None else x.group())
        
        table["Additions"] = table["Additions"].astype(str)
        table["Deposits/ Additions Description"] = table.apply(lambda x: x["Deposits/ Additions Description"].replace(x["Additions"], ""),axis=1)
        
        table["Additions"] = table["Additions"].apply(lambda x: float(x) 
                                                      if x != "None"
                                                      else np.nan)
        
        tables_2[key] = table

    elif len(tables[key].columns) == 5:
        
        table = tables[key]
        table = table.iloc[1:, :] #delete first row
        table = table.rename(columns={"Unnamed: 0":"Date"})
        
        while "Totals" in table.Date.values[len(table) - 1] or "Ending" in table.Date.values[len(table) - 1]:
            table = table.drop(table.index[len(table)-1])

        tables_2[key] = table
        
    else:
        table = tables[key]
        
        patron = "(\d*|\d{1,3}(,\d{3})*)(\.\d+)"
        regex_number_extractor = re.compile(patron)
        table["Additions/Withdrawals"] = table["Deposits/ Additions Withdrawals/ Subtractions Description"].apply(lambda x: regex_number_extractor.search(x))
        table["Additions/Withdrawals"] = table["Additions/Withdrawals"].apply(lambda x: x if x is None else x.group())
        
        table["Additions/Withdrawals"] = table["Additions/Withdrawals"].astype(str)
        table["Additions/Withdrawals"]  = table["Additions/Withdrawals"].apply(lambda x: x.replace(",", ""))
        table["Additions/Withdrawals"] = table["Additions/Withdrawals"].apply(lambda x: float(x) 
                                                                              if x != "None"
                                                                              else np.nan)
        
        tables_2[key] = table
```
